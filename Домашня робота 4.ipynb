## 1 Від'ємний біноміальний розподіл

```python
y1 = nbinom.pmf(x1, n, p)
plt.stem(x1, y1)
plt.show()
```

## 3 Різні параметри

### 3.1 Біноміальний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин біноміального розподілу з параметрами:
1. $n = 15, p = 0.8$
2. $n = 15, p = 0.2$
3. $n = 20, p = 0.4$
4. $n = 20, p = 0.6$
5. $n = 7, p = 0.5$
6. $n = 25, p = 0.5$
7. $n = 100, p = 0.5$

```python
from scipy.stats import binom

def f3(n, p):
    x = np.arange(-0.5, n+1, 0.5)
    y = binom.pmf(x, n, p)
    plt.stem(x, y)
    plt.show()
    y = binom.cdf(x, n, p)
    plt.plot(x, y)
    plt.show()
    bn_rvs = binom.rvs(n, p, size=10000)
    plt.hist(bn_rvs)
    plt.show()
    
f3(n=15, p=0.8)
f3(n=15, p=0.2)
f3(n=20, p=0.4)
f3(n=20, p=0.6)
f3(n=7, p=0.5)
f3(n=25, p=0.5)
f3(n=100, p=0.5)
```
